Ok, let's get started.

```python
from __future__ import division # python 2.7
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
plt.style.use('ggplot')
import os
import glob
import datashader as ds
import datashader.transfer_functions as tf
```

## Create the dataframe

```python
# Read files into a dataframe

# FYI this takes a minute...
    
# Assuming file directories are placed in directory ./t-drive-sample/
path = r'./t-drive-sample-part*/0*/'
taxi_files = glob.glob(os.path.join(path, "*.txt"))   
column_names = ['taxi_id', 'date_time', 'longitude', 'latitude']
df_master = pd.concat(pd.read_csv(f, names=column_names) for f in taxi_files)  # glue all data into the dataframe

df_master['date_time'] = pd.to_datetime(df_master.date_time) # Correct the type in date_time column
```

```python
df = df_master.copy() # Allows you to 'restart' the worksheet without waiting to recreate dataframe
```

## Distribution of time intervals

Next let's look at the distribution of time intervals. For comparison, here are the graphs we'd like to reproduce. ([source](http://research.microsoft.com/pubs/152883/User_guide_T-drive.pdf))

![title](img-paper.jpg)

```python
# Calculate the intervals in minutes. 
df_time_diffs = df['date_time'].diff().astype('timedelta64[m]')

# Histogram
df_time_hist = df_time_diffs.hist(normed=True, range=(0,12))
df_time_hist.set_ylabel("Frequency")
df_time_hist.set_xlabel("Time Intervals (Minutes)")
plt.show()
```

Interestingly, the same shape appears at two levels: minutes & seconds. This reflects the choice of common intervals 1sec, 5sec, 1min, 5min. To see this, you can run the code after swapping *m* to *s* in *'timedelta64[m]'*. 


# (1.B) Graph distance intervals. 

## Calculate GPS distances

Next we'd like to compute incremental distances between GPS points. This dataset is large enough that ordinary python code might run too slow. There are a number of algorithms and tools for computing the distance between two GPS points. In fact, I tested several tools from python libraries like geopy (vincenty) and geographiclib (Geodesic.WGS84.Inverse). I found them to be far too slow to use (they'd take at least a day to finish!). Instead, I chose a simple numerical formula written in vectorized python (numpy) for speed. 

```python
# Use numpy functions for the distance formula

def gps_dist(a, b, c, d):    
    '''Compute the distance (in meters) between two gps locations. Input is assumed to be a = longitude, b = latitude, etc.'''
    r = 0.0174533  # 1 degree in radians
    return 2 * 6371000 * np.arcsin( np.sqrt( # https://en.wikipedia.org/wiki/Haversine_formula
        np.sin(r*(d - b)/2.0)**2 + np.cos(r*b) * np.cos(r*d) * np.sin(r*(c - a)/2.0)**2))
```

Next we'd like to reproduce Figure 1 (b).

![title](img-paper.jpg)

To do this, compute the row by row distance intervals, and draw its histogram.

```python
df['dist'] = gps_dist(df.longitude,df.latitude, df.longitude.shift(),df.latitude.shift())  # shifted array
df.dist.hist(range=(0,8000), normed=True)
plt.show()
```
